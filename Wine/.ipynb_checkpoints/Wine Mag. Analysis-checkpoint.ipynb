### Bang for your buck

Judging by the results above, it seems like when it comes to spending money on wine, you get diminishing returns pretty quickly. To analyze this issue further, we can use local linear regression to obtain an estimate of the average marginal gain in wine quality for each additional dollar spent. In other words, we can get an estimate for the average "bang for each additional buck".

We'll fit the following linear model:

$$ \text{Points} \approx \hat{\beta_0} + \hat{\beta_1} \text{Price}$$

in small ranges of prices (think \\$0 to \\$15, \\$15 to \\$30, etc.) and then analyze how the estimated coefficient $\hat{\beta_1}$ varies.

```python
def ROI(df, p_low, p_high):
    """
    Returns the marginal gain in points per dollar spent in a certain range of prices,
    from p_low to p_high.
    """
    df_rel = df[(df['price'] >= p_low) & (df['price'] <= p_high)]
    ols = smf.ols('points ~ price', data=df_rel).fit()
    return ols.params[1]
```

```python
increments = [(i*20,(i+1)*20) for i in range(20)]
ROIs = [ROI(df, inc[0], inc[1]) for inc in increments]
plt.title('Bang for Each Additional Buck ($20 increments)')
plt.xlabel('Price range (lower end)')
plt.ylabel('Marginal gain')
plt.plot([inc[0] for inc in increments], ROIs)
```

So we see that the estimated coefficient $\hat{\beta_1}$ decreases pretty quickly as the price of wine increases. In fact, once we look at wines that cost \\$100+, we find that $\hat{\beta_1}$ hits zero and even becomes negative at points. This suggests that you get less bang for each additional buck the more you spend on wine. 

But how reliable are these estimates? Let's take a look at the spread of prices within each of the same windows of prices and analyze that.

```python
def avg(df, p_low, p_high):
    df_rel = df[(df['price'] >= p_low) & (df['price'] <= p_high)]
    return df_rel['points'].mean()

def std(df, p_low, p_high):
    df_rel = df[(df['price'] >= p_low) & (df['price'] <= p_high)]
    return df_rel['points'].std()
```

```python
increments = [(i*20,(i+1)*20) for i in range(15)]
stds = [std(df, inc[0], inc[1]) for inc in increments]
plt.title('Point Variation in Different Price Ranges ($20 increments)')
plt.xlabel('Price range (lower end)')
plt.ylabel('Standard deviation')
plt.plot([inc[0] for inc in increments], stds)
```

It seems like as the price range increases, the range of quality increases as well. For example, in the cheapest price range from \\$0 to \\$15 the standard deviation of points is around 2.1, whereas in the price range from \\$45 to \\$60 the standard devation of points is around 2.75. At higher and higher prices, the estimates become pretty noisy, however.

```python
increments = [(i*20,(i+1)*20) for i in range(15)]
avgs = [avg(df, inc[0], inc[1]) for inc in increments]
stds = [std(df, inc[0], inc[1]) for inc in increments]
plt.title('Average Score per Price Range ($20 increments)')
plt.xlabel('Price range (lower end)')
plt.ylabel('Average score')
plt.errorbar([inc[0] for inc in increments], avgs, yerr=stds)
```

This is pretty illuminating. First, it seems as though the average wine quality increases sublinearly as the price range increases, supporting the findings from the local regression estimates above regarding diminishing returns to price. Second, it seems as though the variation in quality within each price range is greater than the variation between adjacent price ranges.

```python
increments = [(i*5,(i+1)*5) for i in range(50)]
avgs = [avg(df, inc[0], inc[1]) for inc in increments]
stds = [std(df, inc[0], inc[1]) for inc in increments]
plt.title('Average Score per Price Range ($5 increments)')
plt.xlabel('Price range (lower end)')
plt.ylabel('Average score')
plt.errorbar([inc[0] for inc in increments], avgs, yerr=stds)
```

This is the same curve as the previous, but just with more granular price ranges. We can clearly see the sublinear shape of the curve at this more granular level.

On the other hand, we can analyze the inverse relationship (between wine quality and average price) by measuring the average price of a bottle of wine at each score from 80 to 100. This provides a sense of how much we would expect to pay for a typical bottle of wine at a certain quality (score).

```python
def avg_price(df, score_low, score_high):
    df_rel = df[(df['points'] >= score_low) & (df['points'] <= score_high)]
    return df_rel['price'].mean()

def median_price(df, score_low, score_high):
    df_rel = df[(df['points'] >= score_low) & (df['points'] <= score_high)]
    return df_rel['price'].median()

def quantile_price(df, score_low, score_high):
    df_rel = df[(df['points'] >= score_low) & (df['points'] <= score_high)]
    return df_rel['price'].quantile(.75) - df_rel['price'].quantile(.25)

def std_price(df, score_low, score_high):
    df_rel = df[(df['points'] >= score_low) & (df['points'] <= score_high)]
    return df_rel['price'].std()
```

```python
increments = [(80 + i, 80 + i + 1) for i in range(20)]
avgs = [avg_price(df, inc[0], inc[1]) for inc in increments]
stds = [std_price(df, inc[0], inc[1]) for inc in increments]
plt.title('Average Price per Score')
plt.xlabel('Score')
plt.ylabel('Average price')
plt.errorbar([inc[0] for inc in increments], avgs, yerr=stds)
```

It seems that average prices grow fairly slowly for bottles of wine with a score of 90 and under, yet prices blow up pretty quickly for the best bottles of wines. Moreover, the variation in prices also increases dramatically as the quality of wine increases. This suggests that low quality bottles of wine can be had at cheap prices consistently, whereas high quality bottles of wine can either be (relatively) cheap or very expensive.

```python
increments = [(80 + i, 80 + i + 1) for i in range(20)]
medians = [median_price(df, inc[0], inc[1]) for inc in increments]
stds = [quantile_price(df, inc[0], inc[1]) for inc in increments]
plt.title('Median Price per Score')
plt.xlabel('Score')
plt.ylabel('Median price')
plt.errorbar([inc[0] for inc in increments], avgs, yerr=stds)
```

This is almost the same curve as the previous, but now considering median prices as opposed to mean prices. In addition, the error bars now indicate the difference in the 75th and 25th quantiles of prices, as opposed to the standard deviation in prices. All in all, this median curve looks very similar to the mean curve, and we see that median prices grow a bit more slowly at the top after accounting for ultra expensive outlier wines.

```python
increments = [(80 + i, 80 + i + 1) for i in range(20)]
medians = [median_price(df, inc[0], inc[1]) for inc in increments]
avgs = [avg_price(df, inc[0], inc[1]) for inc in increments]
plt.plot([inc[0] for inc in increments], avgs, label='Average')
plt.plot([inc[0] for inc in increments], medians, label='Median')
plt.title("Average vs. Median Price at Quality Scores")
plt.xlabel("Score")
plt.ylabel("Price")
plt.legend()
plt.show()
```

Here we can more clearly see the relationship between mean and median prices of wines at each quality. The mean and median prices of wines are very similar until a score of about 90, at which point they diverge. This can be attributed to the ultra expensive outliers we saw before.

### Takeaways

In all, the most important takeaway from this section is that when it comes to buying wine, there are noticeable diminishing returns to price. At the cheaper end of the spectrum (say, \\$20 and under), each additional dollar spent has a meaningful positive effect on the quality of wine your money can buy. However, once you start considering more expensive wines (say, \\$100 and up), increases in quality with each additional dollar spent become less and less detectable. At the highest end of the spectrum (say, \\$500 and up) there is almost no correlation between marginal price and quality anymore, and the variation in quality is very high. Some of these ultra expensive are of high quality as you'd expect, yet some are of surprisingly low quality.

In fact, you can buy a very solid bottle of wine for \\$25 and under. The difference in quality between a typical \\$25 bottle of wine and a typical \\$50 bottle of wine is much smaller than the difference in quality between the best \\$25 bottle and the worst \\$25 bottle. 

## Taking a Closer Look

The analysis so far has been interesting, but let's take it a step further and analyze wines at a more granular level. In particular, let's analyze wines based on country of origin and grape varietals.

```python
common_countries = country_counts[:5]
#common_countries = [c for c in countries if str(c) != 'nan' and country_counts[c] >= 6000]
```

```python
def avg_country(df, p_low, p_high, country):
    df_rel = df.loc[df['country'] == country]
    df_rel = df_rel[(df['price'] >= p_low) & (df['price'] <= p_high)]
    return df_rel['points'].mean()
def std_country(df, p_low, p_high, country):
    df_rel = df.loc[df['country'] == country]
    df_rel = df_rel[(df['price'] >= p_low) & (df['price'] <= p_high)]
    return df_rel['points'].mean()

def avg_price_country(df, score_low, score_high, country):
    df_rel = df.loc[df['country'] == country]
    df_rel = df_rel[(df['points'] >= score_low) & (df['points'] <= score_high)]
    return df_rel['price'].median()
def std_price_country(df, score_low, score_high, country):
    df_rel = df.loc[df['country'] == country]
    df_rel = df[(df['points'] >= score_low) & (df['points'] <= score_high)]
    return df_rel['price'].std()
```

```python
def avg_varietal(df, p_low, p_high, varietal):
    df_rel = df.loc[df['variety'] == varietal]
    df_rel = df_rel[(df['price'] >= p_low) & (df['price'] <= p_high)]
    return df_rel['points'].mean()
def std_varietal(df, p_low, p_high, varietal):
    df_rel = df.loc[df['variety'] == varietal]
    df_rel = df_rel[(df['price'] >= p_low) & (df['price'] <= p_high)]
    return df_rel['points'].mean()

def avg_price_varietal(df, score_low, score_high, varietal):
    df_rel = df.loc[df['variety'] == varietal]
    df_rel = df_rel[(df['points'] >= score_low) & (df['points'] <= score_high)]
    return df_rel['price'].median()
def std_price_varietal(df, score_low, score_high, varietal):
    df_rel = df.loc[df['variety'] == varietal]
    df_rel = df[(df['points'] >= score_low) & (df['points'] <= score_high)]
    return df_rel['price'].std()
```

```python
def avg_countries(df, countries, bandwidth):
    increments = [(i*10,(i+1)*10) for i in range(15)]
    country_avgs = {}
    for country in countries:
        country_avgs[country] = [avg_country(df, inc[0], inc[1], country) for inc in increments]
    for country in countries:
        plt.plot([inc[0] for inc in increments], country_avgs[country], label=country)
    plt.title('Average Score for wines ($' + str(bandwidth) + ' increments)')
    plt.xlabel('Price range (lower end)')
    plt.ylabel('Average score')
    plt.legend()
    plt.show()
```

```python
avg_countries(df, common_countries, 10)
```

```python
def avg_price_countries(df, countries):
    increments = [(80 + i, 80 + (i + 1)) for i in range(15)]
    #increments = [(i, i+1) for i in range(87, 93)]
    country_price_avgs = {}
    for country in countries:
        country_price_avgs[country] = [avg_price_country(df, inc[0], inc[1], country) for inc in increments]
    for country in countries:
        plt.plot([inc[0] for inc in increments], country_price_avgs[country], label=country)
    plt.title('Average Price per Score')
    plt.xlabel('Score')
    plt.ylabel('Average price')
    plt.legend()
    plt.show()
```
